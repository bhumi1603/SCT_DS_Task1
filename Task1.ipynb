```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("age_distribution.csv")

# Set style
sns.set_theme(style="whitegrid")

# ✅ 1. Bar Chart Visualization
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="Age Group", y="Population (millions)", hue="Age Group", data=df, 
                 palette={"0 - 20 Years": "yellow", "21 - 64 Years": "blue", "65+ Years": "pink"},
                 legend=False)

# Add labels
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Population (millions)", fontsize=12)
plt.title("Population Distribution by Age in 2022", fontsize=14)
plt.grid(axis='y', linestyle="--", alpha=0.7)

# Show values on bars
for i, v in enumerate(df["Population (millions)"]):
    plt.text(i, v + 20, f"{v}M", ha='center', fontsize=12, fontweight='bold')

plt.show()
```
